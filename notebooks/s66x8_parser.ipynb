```python
import hdnntools as hdt
import numpy as np
import os
```

```python
fdir = '/home/jujuman/Research/extensibility_test_sets/s66x8/structures/'
sdir = '/home/jujuman/Research/extensibility_test_sets/s66x8/xyzfiles/'
```

```python
files = os.listdir(fdir)
```

```python
files.sort()
```

```python
Nd = int(len(files)/8)
for i in range(Nd):
    print('Gathering data:')
    X = []
    for file in files[i*8:i*8+8]:
        Xi,S,Na,c = hdt.readxyz2(fdir+file)
        print(S)
        X.append(Xi)
    X = np.concatenate(X)
    hdt.writexyzfile(sdir+'s66ds_'+str(i).zfill(2)+'.xyz',xyz=X,typ=S)
        
```
