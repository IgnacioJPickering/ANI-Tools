```python
import hdnntools as hdt
import pyanitools as pyt
import os
import numpy as np
```

```python
wdir = '/home/jujuman/Seagate2TB/Research/GDB-11-wB97X-6-31gd/dnntsgdb11_10/'
```

```python
ipt_files =  [f for f in os.listdir(wdir) if '.ipt' in f]
dat_files =  [f for f in os.listdir(wdir+'data/') if '.dat' in f]
```

```python
d = dict()
for i in ipt_files:
    d.update({i.split('.')[0]:i})
```

```python
dp = pyt.datapacker(wdir+'new_h5_gdb10_ts.h5')
```

```python
for i in dat_files:
    k = i.rsplit('_',1)[0]
    f = open(wdir+d[k],'r').read()
    smiles = f.split('#Smiles: ')[1].split('\n')[0]
    print(smiles)
    datal = hdt.readncdat(wdir+'data/'+i,type=np.float64)
    
    data = {'coordinates':np.array(datal[0],dtype=np.float32),
            'species':datal[1],
            'energies':datal[2],
            'smiles':[smiles],
           }
    
    loc = '/gdb10_test/'+k.split('-')[0]+'-'+str(int(k.split('-')[1])).zfill(3)
    print(loc)
    dp.store_data(loc, **data)
dp.cleanup()
```
