```python
aa_l = ['A','V','I','L','M','F','Y','W','S','T','N','Q','C','G','P','R','H','K','D','E']

fasta = []
#for i,aa in enumerate(aa_l):
#    s = ">AKI:A|PDBID|CHAIN|SEQUENCE\n"+aa
#    fasta.append(('aachain_'+aa+'-'+str(i).zfill(4),s))
    
#for i,aai in enumerate(aa_l):
#    for j,aaj in enumerate(aa_l):
#        s = ">AKI:A|PDBID|CHAIN|SEQUENCE\n"+aai+aaj
#        fasta.append(('aachain_'+aai+aaj+'-'+str(i).zfill(2)+str(j).zfill(2),s))
        
for i,aai in enumerate(aa_l):
    for j,aaj in enumerate(aa_l):
        for k,aak in enumerate(aa_l):
            s = ">AKI:A|PDBID|CHAIN|SEQUENCE\n"+aai+aaj+aak
            fasta.append(('aa_'+aai+aaj+aak+'-'+str(i).zfill(2)+str(j).zfill(2)+str(k).zfill(2),s))
```

```python
for i,f in enumerate(fasta):
    m = Chem.MolFromFASTA(f[1])
    at = [str(j.GetSymbol()) for j in m.GetAtoms()]
    if len(set(at).intersection(set(['S']))) > 0:
        print(at)
        m = Chem.AddHs(m)
        #AllChem.EmbedMolecule(m, useRandomCoords=True)
    
        cids = AllChem.EmbedMultipleConfs(m, numConfs=1)
        for cid in cids:
            _ = AllChem.MMFFOptimizeMolecule(m, confId=cid)

        rmslist = []
        AllChem.AlignMolConformers(m, RMSlist=rmslist)
        #E = act.detect_unique_rdkitconfs(m, cids)
        X, S = pya.__convert_rdkitconfs_to_nparr__(m)
        print(f[0],len(S),X.shape)
        hdt.writexyzfile(wdir+'XYZ/'+f[0]+'.xyz',X,S)
        for j,x in enumerate(X):
            hdt.write_rcdb_input(x,S,j,wdir,f[0], 5, 'wb97x/6-31g*','300.0',fill=2,comment='#FAST:'+f[1])
```
